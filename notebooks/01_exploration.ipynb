# Copper Macro Research Model
### Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='darkgrid')
%matplotlib inline

master = pd.read_csv('../data/processed/master.csv', parse_dates=['date'], index_col='date')
print(master.shape)
master.tail()
```

```python
fig, axes = plt.subplots(3, 2, figsize=(14, 10))
fig.suptitle('Copper Macro Dashboard', fontsize=14)

master['lme_close'].plot(ax=axes[0,0], title='LME Copper (USD/lb)')
master['usdcny_close'].plot(ax=axes[0,1], title='USD/CNY')
master['china_manufacturing_pmi'].plot(ax=axes[1,0], title='China Manufacturing PMI')
master['comex_copper_close_usd'].plot(ax=axes[1,1], title='COMEX Copper (SHFE proxy)')
master['chile_copper_production_mt'].plot(ax=axes[2,0], title='Chile Copper Production (MT)')
master['lme_volume'].plot(ax=axes[2,1], title='LME Volume')

plt.tight_layout()
plt.show()
```

```python
corr_cols = ['lme_close','usdcny_close','china_manufacturing_pmi',
             'comex_copper_close_usd','chile_copper_production_mt']
corr = master[corr_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```
